```python
import os, librosa
import numpy as np
import scipy.signal, torch, math
import matplotlib.pyplot as plt
```

```python
wav_file = '/media/nas_mount/hemant/ner/asr_wm/gaurav.wav'
# loading the audio
y, sr = librosa.load(wav_file,sr = 8000)
# waveform plot of the audio
plt.plot(y)
```

```python
n_fft = int(8000*0.02)
win_length = n_fft
hop_length = int(8000*0.01)
D = librosa.stft(y, n_fft=n_fft, hop_length=hop_length,
                 win_length=win_length)
```

```python
spect, phase = librosa.magphase(D)
spect = np.log1p(spect)
spect = torch.FloatTensor(spect)

# plotting the spectrogram image.
plt.imshow(spect)
```
